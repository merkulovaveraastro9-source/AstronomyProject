# Dark Mass

```python
import numpy as np
from h5py import File
import matplotlib.pyplot as mpl
import requests

baseUrl = 'http://www.tng-project.org/api/'
headers = {"api-key":"fe476e8c1fe04ea8f764a361b591220a"}
h0 = 0.6774
m_dm = 3.64755609660833*10**(-5)

def get(path, params=None):
    # make HTTP GET request to path
    r = requests.get(path, params=params, headers=headers)

    # raise exception if response code is not HTTP SUCCESS (200)
    r.raise_for_status()

    if r.headers['content-type'] == 'application/json':
        return r.json() # parse json responses automatically

    if 'content-disposition' in r.headers:
        filename = r.headers['content-disposition'].split("filename=")[1]
        with open(filename, 'wb') as f:
            f.write(r.content)
        return filename # return the filename string

    return r
def GetCutout(baseUrl,number_snap,ID):
    simulation=baseUrl + "TNG50-1-Dark/"
    snapshot = get(get(simulation)['snapshots'])[number_snap]['url']
    
    redshift = get(snapshot)['redshift']
    
   
    subhalo = get(get(snapshot)['subhalos'])['results'][ID]['url']
    cutout = get(subhalo + "cutout.hdf5",{'dm':'Coordinates'})
    scale_factor = 1.0 / (1+redshift)
    return cutout,subhalo,scale_factor
    
def Dark_Mass(cutout,subhalo,scale_factor):
    with File(cutout,'r') as cut:
        x = cut['PartType1']['Coordinates'][:,0] - get(subhalo)['pos_x']
        y = cut['PartType1']['Coordinates'][:,1] - get(subhalo)['pos_y']
        z = cut['PartType1']['Coordinates'][:,2] - get(subhalo)['pos_z']
        radius = np.sqrt(x**2 + y**2 + z**2) * scale_factor*h0
        Mass= [ m_dm*h0*10**10 for i in range(len(radius))]
        hist, bins = np.histogram(radius, bins=500,weights=Mass,density = False)
        Masses = np.cumsum(hist)
        Rad = np.zeros(len(hist))
        for i in range(len(hist)):
            Rad[i] = (bins[i] + bins[i+1]) / 2
        dR = np.zeros(len(hist))
        for i in range(len(hist)):
            dR[i] = bins[i+1] - bins[i]
        
        mpl.plot(Rad,Masses)
        
        mpl.show()
    return Masses,Rad, hist, dR
        
        
        
```

# Plotnost and curved Mass

```python
import numpy as np
import scipy
from scipy.optimize import curve_fit
import matplotlib.pyplot as mpl
def Plotnost_Dark_Mass(Mass,Rad, dR):
    Plotnost  = Mass / (4 * np.pi * Rad**2 * dR)
    mpl.plot(Rad,Plotnost)
    mpl.show()
    return Plotnost, Rad
def NFW(Rad, r0,plot0):
    return plot0/((Rad/r0) * (1+Rad/r0)**2)
def curve_plotnost(Plotnost,Rad):
    params,conv= curve_fit(NFW,Rad,Plotnost, p0 = [10**9,70], maxfev = 5000)
    r0, plot0 = params
    
    print(plot0,r0)
    return plot0,r0
def curve__func_plot(Rad,plot0,r0,Plotnost):
    fig,ax = mpl.subplots()
    ax.plot(Rad,Plotnost,label = 'exp')
    ax.plot(Rad,  plot0/((Rad/r0) * (1+Rad/r0)**2),label = 'curved')
    ax.set_xlabel('Radius')
    ax.set_ylabel('Plotnost')
    ax.legend()
def curve__func_mass(Rad,plot0,r0,Mass):
    fig,ax = mpl.subplots()
    ax.plot(Rad,Mass,label = 'exp')
    ax.plot(Rad,  4 * np.pi * plot0 * r0**3 * (np.log((r0 + Rad)/r0) - Rad / (r0 + Rad)),label = 'curved')
    ax.set_xlabel('Radius')
    ax.set_ylabel('Mass')
    ax.legend()
```

```python
cutout,subhalo,scale_factor = GetCutout(baseUrl,99,1)
Masses,Rad,Mass,dR  = Dark_Mass(cutout,subhalo,scale_factor)
Plotnost,Rad= Plotnost_Dark_Mass(Mass,Rad,dR)
plot0,r0 = curve_plotnost(Plotnost,Rad)
curve__func_plot(Rad,plot0,r0,Plotnost)
```

```python
cutout,subhalo,scale_factor = GetCutout(baseUrl,99,8)
Masses,Rad,Mass,dR  = Dark_Mass(cutout,subhalo,scale_factor)
Plotnost,Rad= Plotnost_Dark_Mass(Mass,Rad,dR)
plot0,r0 = curve_plotnost(Plotnost,Rad)
curve__func_plot(Rad,plot0,r0,Plotnost)
curve__func_mass(Rad,plot0,r0,Masses)
```
