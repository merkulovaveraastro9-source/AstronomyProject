# Dark Matter

```python
import numpy as np
from h5py import File
import matplotlib.pyplot as plt
import requests

baseUrl = 'http://www.tng-project.org/api/'
headers = {"api-key":"fe476e8c1fe04ea8f764a361b591220a"}
h0 = 0.6774
m_dm = 3.64755609660833*10**(-5)

def get(path, params=None):
    # make HTTP GET request to path
    r = requests.get(path, params=params, headers=headers)

    # raise exception if response code is not HTTP SUCCESS (200)
    r.raise_for_status()

    if r.headers['content-type'] == 'application/json':
        return r.json() # parse json responses automatically

    if 'content-disposition' in r.headers:
        filename = r.headers['content-disposition'].split("filename=")[1]
        with open(filename, 'wb') as f:
            f.write(r.content)
        return filename # return the filename string

    return r
def GetCutout(baseUrl,number_snap,id, n_time):
    simulation=baseUrl + "TNG50-1/"
    snapshot = get(get(simulation)['snapshots'])[number_snap]['url']
    
    redshift = get(snapshot)['redshift']
    
   
    subhalo = snapshot + f"subhalos/{id}/mwm31s_cutouts/"
    cutout = get(subhalo + f"snap_{n_time}.hdf5")
    scale_factor = 1.0 / (1+redshift)
    return cutout, scale_factor
    
def Dark_Mass(cutout,scale_factor):
    with File(cutout,'r') as cut:
        x = cut['PartType1']['RotatedCoordinates'][:,0]
        y = cut['PartType1']['RotatedCoordinates'][:,1]
        z = cut['PartType1']['RotatedCoordinates'][:,2]
        radius = np.sqrt(x**2 + y**2 + z**2) * scale_factor*h0
        Mass= [ m_dm*h0*10**10 for i in range(len(radius))]
        hist, bins = np.histogram(radius, bins=500,weights=Mass,density = False)
        Masses = np.cumsum(hist)
        Rad = np.zeros(len(hist))
        for i in range(len(hist)):
            Rad[i] = (bins[i] + bins[i+1]) / 2
        dR = np.zeros(len(hist))
        for i in range(len(hist)):
            dR[i] = bins[i+1] - bins[i]
        
        plt.plot(Rad,Masses)
        plt.xlabel(r'$R [kpc]$')
        plt.ylabel(r'$M [{M}_{sun}]$')
        
        plt.show()
    return Masses,Rad, hist, dR
def Dark_Mass_Density(Mass,Rad, dR):
    Density  = Mass / (4 * np.pi * Rad**2 * dR)
    plt.plot(Rad,Density)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel(r'$R [kpc]$')
    plt.ylabel(r'$\rho[{M}_{sun}/{kpc}^3]$')
    
    plt.show()
    return Density, Rad        
        
        
```

## NFW profile

```python
import numpy as np
import scipy
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import matplotlib.scale as mscale

def NFW(Rad, r0,Rho0):
    return Rho0/((Rad/r0) * (1+Rad/r0)**2)
def curve_density(Density,Rad):
    params,conv= curve_fit(NFW,Rad,Density, p0 = [70,10**9], maxfev = 100000000)
    r0, Rho0 = params
    
    print(Rho0,r0)
    return Rho0,r0
def curve_func_density(Rad,Rho0,r0,Density):
    fig,ax = plt.subplots()
    ax.plot(Rad,Density,label = 'exp')
    ax.plot(Rad,  Rho0/((Rad/r0) * (1+Rad/r0)**2),label = 'curved')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel(r'$R [kpc]$')
    ax.set_ylabel(r'$\rho[{M}_{sun}/{kpc}^3]$')
    ax.legend()
def curve_func_mass(Rad,Rho,r0,Mass):
    fig,ax = plt.subplots()
    ax.plot(Rad,Mass,label = 'exp')
    ax.plot(Rad,  4 * np.pi * Rho0 * r0**3 * (np.log((r0 + Rad)/r0) - Rad / (r0 + Rad)),label = 'curved')
    ax.set_xlabel(r'$R [kpc]$')
    ax.set_ylabel(r'$M [{M}_{sun}]$')
    ax.legend()
```

```python
cutout,subhalo,scale_factor = GetCutout(baseUrl,99,1)
Masses,Rad,Mass,dR  = Dark_Mass(cutout,subhalo,scale_factor)
Density,Rad= Dark_Mass_Density(Mass,Rad,dR)
Rho0,r0 = curve_density(Density,Rad)
curve_func_density(Rad,Rho0,r0,Density)
```

## Einasto profile

```python
import numpy as np
import scipy
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import matplotlib.scale as mscale

def Einasto(Rad,Rho0E, A, alpha):
    return Rho0E * np.exp(A * Rad**alpha)
def curve_density_Einasto(Density,Rad):
    params,conv= curve_fit(Einasto,Rad,Density,p0 = [10**6,-10,1.4],maxfev = 100000000)
    Rho0E, A, alpha = params
    
    print(Rho0E, A, alpha)
    return Rho0E, A, alpha
def curve_func_density_Einasto(Rad,Rho0E, A, alpha,Density):
    fig,ax = plt.subplots()
    ax.plot(Rad,Density,label = 'exp')
    ax.plot(Rad,Rho0E * np.exp(A * Rad**alpha)  ,label = 'curved')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel(r'$R [kpc]$')
    ax.set_ylabel(r'$\rho[{M}_{sun}/{kpc}^3]$')
    ax.legend()
```

```python
import numpy as np
from h5py import File
import matplotlib.pyplot as plt
import requests

baseUrl = 'http://www.tng-project.org/api/'
headers = {"api-key":"fe476e8c1fe04ea8f764a361b591220a"}
h0 = 0.6774
def get(path, params=None):
    # make HTTP GET request to path
    r = requests.get(path, params=params, headers=headers)

    # raise exception if response code is not HTTP SUCCESS (200)
    r.raise_for_status()

    if r.headers['content-type'] == 'application/json':
        return r.json() # parse json responses automatically

    if 'content-disposition' in r.headers:
        filename = r.headers['content-disposition'].split("filename=")[1]
        with open(filename, 'wb') as f:
            f.write(r.content)
        return filename # return the filename string

    return r

def Barion_Mass(cutout,scale_factor):
    with File(cutout,'r') as cut:
        x = cut['PartType4']['RotatedCoordinates'][:,0]
        y = cut['PartType4']['RotatedCoordinates'][:,1]
        Masses = cut['PartType4']['Masses'][:]
        Masses *= (10**10 * h0)
        radius = np.sqrt(x**2 + y**2) * scale_factor*h0
        hist, bins = np.histogram(radius, bins=50,weights=Masses,density = False)
        Mass = np.cumsum(hist)
        Rad = np.zeros(len(hist))
        for i in range(len(hist)):
            Rad[i] = (bins[i] + bins[i+1]) / 2
        dR = np.zeros(len(hist))
        for i in range(len(hist)):
            dR[i] = bins[i+1] - bins[i]
        plt.plot(Rad,Mass)
        plt.xlabel(r'$R [kpc]$')
        plt.ylabel(r'$M [{M}_{sun}]$')
        
        plt.show()
    return Mass,Rad, hist, dR
def Barion_Mass_Density(Mass,Rad, dR):
    Density  = Mass / (2*np.pi * Rad * dR)
    plt.plot(Rad,Density)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel(r'$R [kpc]$')
    plt.ylabel(r'$\sigma[{M}_{sun}/{kpc}^2]$')
    
    plt.show()
    return Density, Rad
```

```python
cutout,subhalo,scale_factor = GetCutout(baseUrl,99,419618,11)
Mass,Rad,Masses,dR  = Barion_Mass(cutout,subhalo,scale_factor)
Density,Rad= Barion_Mass_Density(Masses,Rad,dR)
```

```python
import numpy as np
import scipy
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import matplotlib.scale as mscale

def ExpProfile(Rad, alpha, A, sigma0):
    return sigma0 * np.exp(A * Rad**alpha)
def curve_density_Exponencial(Density,Rad):
    params,conv= curve_fit(ExpProfile,Rad,Density,p0 = [0.5,-0.5,10**9],maxfev = 100000000)
    alpha, A, sigma0 = params
    
    print(alpha, A, sigma0)
    return alpha, A, sigma0
def curve_func_density_Exponencial(Rad,alpha, A, sigma0,Density):
    fig,ax = plt.subplots()
    ax.plot(Rad,Density,label = 'exp')
    ax.plot(Rad, sigma0 * np.exp(A * Rad**alpha) ,label = 'curved')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel(r'$R [kpc]$')
    ax.set_ylabel(r'$\sigma[{M}_{sun}/{kpc}^2]$')
    ax.legend()
```
