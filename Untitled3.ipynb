# Gas_Mass(Radius)

```python
import numpy as np
from h5py import File
import matplotlib.pyplot as mpl
##from math import sqrt


h0 = 0.6774
def Function_Mass_Radius(baseUrl,number_snap,ID):
    simulation=baseUrl + "TNG50-1/"
    snapshot = get(get(simulation)['snapshots'])[number_snap]['url']
    
    redshift = get(snapshot)['redshift']
    
   
    subhalo = get(get(snapshot)['subhalos'])['results'][ID]['url']
    cutout = get(subhalo + "cutout.hdf5",{'gas':'Coordinates,Masses'})
    scale_factor = 1.0 / (1+redshift)
    with File(cutout,'r') as file:
        
        x = file['PartType0']['Coordinates'][:,0] - get(subhalo)['pos_x']
        y = file['PartType0']['Coordinates'][:,1] - get(subhalo)['pos_y']
        z = file['PartType0']['Coordinates'][:,2] - get(subhalo)['pos_z']
        Mass = file['PartType0']['Masses'][:] * h0 * 10**10
        radius = np.sqrt(x**2 + y**2 + z**2) * scale_factor*h0
        Massive = [[radius[i],Mass[i]] for i in range(len(x))]
        Massive_1 = sorted(Massive, key = lambda item:item[0])
        radius_2 = np.zeros(len(x))
        Mass_1 = np.zeros(len(x))
        for i in range(len(x)):
            radius_2[i] = Massive_1[i][0]
            Mass_1[i] = Massive_1[i][1]
        Integ_Mass = np.zeros(len(x))
        for i  in range(len(x)):
            Integ_Mass[i] =  np.trapezoid(Mass_1[:i],x = radius_2[:i])
        
        
        
        mpl.plot(radius_2,Integ_Mass)
        mpl.legend()
        mpl.xlabel('Radius')
        mpl.ylabel('Mass')
```
